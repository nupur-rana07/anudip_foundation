```python
lst = [10, 20, 4, 45, 99]

first = second = float('-inf')

for num in lst:
    if num > first:
        second = first
        first = num
    elif num > second and num != first:
        second = num

print("Second largest:", second)
```

```python
s = "hello"

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

print(freq)
```

```python
l1 = [1, 2, 3, 4]
l2 = [3, 4, 5, 6]

common = []

for i in l1:
    if i in l2 and i not in common:
        common.append(i)

print(common)
```

```python
n = 153
temp = n
sum = 0
digits = len(str(n))

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10

print(sum == n)
```

```python
lst = [1, 2, 3, 4, 5]
target = 5

pairs = []

for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == target:
            pairs.append((lst[i], lst[j]))

print(pairs)
```

```python
nested = [1, [2, 3], [4, [5, 6]]]

def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i, list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result

print(flatten(nested))
```
