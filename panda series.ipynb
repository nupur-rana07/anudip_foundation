```python
# Step 1: Install & Import (pandas is already installed in Colab, but keeping for safety)
!pip install pandas

import pandas as pd

# Step 2: Create Dataset
marks = [45, 67, 89, 34, 90, 55, 72]
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS', 'Biology', 'History']

# Create Series
s = pd.Series(marks, index=subjects)

# Step 3: Display Original Data
print("===== ORIGINAL DATA =====")
print(s)

# Step 4: Find subjects with marks > 70
print("\n===== SUBJECTS WITH MARKS > 70 =====")
above_70 = s[s > 70]
print(above_70)

# Step 5: Replace marks < 40 with "Fail"
print("\n===== AFTER REPLACING MARKS < 40 WITH 'Fail' =====")
s_updated = s.copy()
s_updated[s_updated < 40] = "Fail"
print(s_updated)

# Step 6: Count subjects above average
print("\n===== ABOVE AVERAGE ANALYSIS =====")
average_marks = s.mean()
count_above_avg = (s > average_marks).sum()

print("Average Marks:", average_marks)
print("Number of subjects above average:", count_above_avg)

# Step 7: Optional - Save Output to CSV in Colab
s_updated.to_csv("marks_output.csv")

print("\nFile 'marks_output.csv' saved successfully!")
```

```python
# Install (optional, already available in Colab)
!pip install pandas

import pandas as pd

# Step 1: Create Dataset
salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']

s = pd.Series(salary, index=emp)

# Step 2: Display Original Data
print("===== ORIGINAL SALARY DATA =====")
print(s)

# Step 3: Increase salary by 10% where salary < 30000
s_updated = s.copy()
s_updated[s_updated < 30000] = s_updated[s_updated < 30000] * 1.10

# Step 4: Decrease salary by 5% where salary > 45000
s_updated[s_updated > 45000] = s_updated[s_updated > 45000] * 0.95

# Step 5: Display Updated Salaries
print("\n===== UPDATED SALARY DATA =====")
print(s_updated)

# Step 6: Total Salary Expense after changes
total_expense = s_updated.sum()

print("\n===== TOTAL SALARY EXPENSE =====")
print("Total Salary Expense:", total_expense)

# Step 7: Save to CSV (optional)
s_updated.to_csv("updated_salary.csv")
print("\nFile 'updated_salary.csv' saved successfully!")
```

```python


import pandas as pd

# Step 1: Create Dataset
data = [10, None, 30, None, 50, 60, None]
s = pd.Series(data)

# Step 2: Display Original Data
print("===== ORIGINAL DATA =====")
print(s)

# Step 3: Count missing values BEFORE filling
missing_before = s.isnull().sum()
print("\nMissing values before filling:", missing_before)

# Step 4: Fill missing values with mean
mean_value = s.mean()
s_filled = s.fillna(mean_value)

print("\n===== DATA AFTER FILLING WITH MEAN =====")
print(s_filled)

# Step 5: Count missing values AFTER filling
missing_after = s_filled.isnull().sum()
print("\nMissing values after filling:", missing_after)

# Step 6: Drop remaining nulls (if any)
s_final = s_filled.dropna()

print("\n===== FINAL DATA AFTER DROPPING NULLS =====")
print(s_final)

# Step 7: Save to CSV (optional)
s_final.to_csv("cleaned_data.csv")
print("\nFile 'cleaned_data.csv' saved successfully!")
```

```python


import pandas as pd

# Step 1: Create Dataset
marks = [95, 82, 67, 58, 73, 49, 88]
students = ['S1','S2','S3','S4','S5','S6','S7']

s = pd.Series(marks, index=students)

# Step 2: Display Original Data
print("===== ORIGINAL MARKS =====")
print(s)

# Step 3: Define grading function
def assign_grade(m):
    if m >= 90:
        return 'A'
    elif m >= 75:
        return 'B'
    elif m >= 60:
        return 'C'
    else:
        return 'D'

# Step 4: Apply grading
grades = s.apply(assign_grade)

print("\n===== ASSIGNED GRADES =====")
print(grades)

# Step 5: Count number of students in each grade
grade_count = grades.value_counts()

print("\n===== GRADE COUNT =====")
print(grade_count)

# Step 6: Save to CSV (optional)
grades.to_csv("student_grades.csv")
print("\nFile 'student_grades.csv' saved successfully!")
```

```python


import pandas as pd

# Step 1: Create Dataset
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# Step 2: Display Original Series
print("===== SERIES s1 =====")
print(s1)

print("\n===== SERIES s2 =====")
print(s2)

# Step 3: Add both Series
result = s1 + s2

print("\n===== ADDITION RESULT (WITH NaN) =====")
print(result)

# Step 4: Replace NaN with 0 and recompute
s1_filled = s1.fillna(0)
s2_filled = s2.fillna(0)

result_fixed = s1_filled.add(s2_filled, fill_value=0)

print("\n===== AFTER HANDLING NaN (RECOMPUTED) =====")
print(result_fixed)

# Step 5: Save result (optional)
result_fixed.to_csv("series_alignment_result.csv")
print("\nFile 'series_alignment_result.csv' saved successfully!")
```

```python
import pandas as pd

# Dataset
scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']

s = pd.Series(scores, index=players)

# 1. Rank players based on scores (highest = rank 1)
ranked = s.rank(ascending=False)
print("Ranking of Players:\n")
print(ranked)

# 2. Display top 3 players
top3 = s.sort_values(ascending=False).head(3)
print("\nTop 3 Players:\n")
print(top3)

# 3. Find player with lowest score
lowest_player = s.idxmin()
lowest_score = s.min()

print("\nPlayer with Lowest Score:")
print(lowest_player, "->", lowest_score)
```

```python
import pandas as pd

# Dataset
names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

# 1. Convert all names to uppercase
upper_names = s.str.upper()
print("Uppercase Names:\n")
print(upper_names)

# 2. Find names containing letter 'a'
names_with_a = s[s.str.contains('a')]
print("\nNames containing 'a':\n")
print(names_with_a)

# 3. Replace 'anuj' with 'Anuj Kumar'
replaced_names = s.replace('anuj', 'Anuj Kumar')
print("\nAfter Replacement:\n")
print(replaced_names)
```

```python
import pandas as pd

# Dataset
data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

# Chained Operations
result = (
    s.fillna(s.median())      # Fill missing with median
     .loc[lambda x: x > 20]   # Filter values > 20
     .sort_values(ascending=False)  # Sort descending
     .reset_index(drop=True)  # Reset index
)

print("Final Result:\n")
print(result)
```

```python
import pandas as pd

# Dataset
sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

s = pd.Series(sales, index=days)

# 1. Best and worst sales day
best_day = s.idxmax()
worst_day = s.idxmin()

print("Best Sales Day:", best_day, "->", s.max())
print("Worst Sales Day:", worst_day, "->", s.min())

# 2. Total and average sales
total_sales = s.sum()
average_sales = s.mean()

print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)

# 3. Apply 20% bonus where sales > 400
bonus_sales = s.apply(lambda x: x * 1.2 if x > 400 else x)

print("\nSales after Bonus:\n")
print(bonus_sales)
```
